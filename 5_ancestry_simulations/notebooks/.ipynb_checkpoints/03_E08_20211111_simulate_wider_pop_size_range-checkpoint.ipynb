### Plot results

```python
parent_sim_dir='/n/data1/hms/dbmi/farhat/nikki/abscessus/0_NOTEBOOKS/014_tree_simulation_msprime/E08_simulate_real_Ne/simulations/04_20211111_wider_range_pop_sizes/'
```

```python
constant_pop_sizes=[3000,8000,14000]
skyline_pop_sizes=[3000,26000,57000]
subpop_sizes=[718, 143, 71, 14]
```

```python
fig, ax= plt.subplots(1, 3, figsize=(20, 5))
ax=ax.flatten()

labels=[str(i) for i in subpop_sizes]

for count, total_pop in enumerate(constant_pop_sizes):   
    print(count)
    print(total_pop)

    mean_ratio_dict={}
    for subpop in subpop_sizes:
        sim_dir_i=os.path.join(parent_sim_dir, 'total_pop_'+str(total_pop), 'subpop_'+str(subpop), 'mut_fold_change_1')
        sim_path_i=os.path.join(sim_dir_i, 'snp_dist_mean_ratios_1')
        with open(sim_path_i, 'rb') as pickle_file:
            sim_data_i = pickle.load(pickle_file)
            mean_ratio_dict[str(subpop)]=sim_data_i
            
    ax[count].boxplot(list(mean_ratio_dict.values()))
    ax[count].axhline(y=0.011, color='r', linestyle='-')
    ax[count].set_yscale("log")
    
    title=r'$N_e^{ T} = $' + str(total_pop)
    ax[count].set_title(title, size=24)
    
    ax[count].set_xticklabels(labels, fontsize=18)
    ax[count].tick_params(axis='y', labelsize=18)

fig.suptitle('Tree Prior: Constant Pop Size', fontsize=30, y=1.1)
fig.text(0.5, -0.05, 'Subpopulation Size '+r'$(N_e^{A})$', ha='center', fontsize=24)
#fig.text(0.09, 0.5, 'clustering metric', va='center', rotation='vertical', fontsize=24)
```

# B. Simulate a range of population sizes with changes in mutation rate over time:

```python
#!sbatch scripts/20211018_test_real_Ne_loopPopSizes.sh
```

```python
parent_sim_dir='/n/data1/hms/dbmi/farhat/nikki/abscessus/0_NOTEBOOKS/014_tree_simulation_msprime/E08_simulate_real_Ne/simulations/03_test_range_pop_sizes/'
```

```python
total_pop_sizes=[3000,8000,14000,26000,57000]
subpop_sizes=[575,718,875]
mut_fold_changes=[1,2,5,10,15,20,25]
```

```python
fig, ax= plt.subplots(3, 5, figsize=(30, 15))
ax=ax.T.flatten()

labels=[str("1/")+str(i) for i in mut_fold_changes]

for count, pair in enumerate(list(itertools.product(total_pop_sizes, subpop_sizes))):   
    total_pop=pair[0]
    subpop=pair[1]
    
    mean_ratio_dict={}
    for i in mut_fold_changes:
        sim_dir_i=os.path.join(parent_sim_dir, 'total_pop_'+str(total_pop), 'subpop_'+str(subpop), 'mut_fold_change_'+str(i))
        sim_path_i=os.path.join(sim_dir_i, 'snp_dist_mean_ratios_'+str(i))
        with open(sim_path_i, 'rb') as pickle_file:
            sim_data_i = pickle.load(pickle_file)
            mean_ratio_dict[str(i)]=sim_data_i
    
    ax[count].boxplot(list(mean_ratio_dict.values()))
    ax[count].axhline(y=0.011, color='r', linestyle='-')
    ax[count].set_yscale("log")
    
    title=r'$N_e^{ T} = $' + str(total_pop)+' , ' + r'$N_e^{A}$' + " = " +str(subpop)
    ax[count].set_title(title, size=24)
    
    ax[count].set_xticklabels(labels, fontsize=16)
    ax[count].tick_params(axis='y', labelsize=16)
    
    fig.tight_layout()
```
