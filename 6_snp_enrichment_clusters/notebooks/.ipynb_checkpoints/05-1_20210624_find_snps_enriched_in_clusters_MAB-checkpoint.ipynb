Now read in the alingment

```python
treemmer_aln=AlignIO.read(masked_snpAln_treemmer_filtered, 'fasta')
```

```python
posn_list=[treemmer_aln[:,p] for p in range(len(treemmer_aln[0].seq))]
```

```python
allele_counts=[[i.count('A'), i.count('C'), i.count('G'), i.count('T')] for i in posn_list]
```

```python
np.shape(allele_counts)
```

```python
minor_alleles_per_SNP_pos = [ list(pd.Series(np.array(SNP_i_allele_count) , index = ['A' , 'C' , 'G' , 'T'])[np.array(SNP_i_allele_count) > 0].sort_values(ascending = True)[:-1].index) for SNP_i_allele_count in allele_counts ]
```

```python
major_alleles_per_SNP_pos = [ list(pd.Series(np.array(SNP_i_allele_count) , index = ['A' , 'C' , 'G' , 'T'])[np.array(SNP_i_allele_count) > 0].sort_values(ascending = True)[-1:].index) for SNP_i_allele_count in allele_counts ]
```

```python
len(minor_alleles_per_SNP_pos)
```

```python
major_allele_list=[i[0] for i in major_alleles_per_SNP_pos]
```

Reformat into a dataframe:

```python
with open('../vars/mab_masked_snp_positions.txt', 'r') as filehandle:
    masked_snp_pos = [pos.rstrip() for pos in filehandle.readlines()]
```

```python
masked_major_minor_alleles = pd.DataFrame(list(zip(masked_snp_pos, major_allele_list, minor_alleles_per_SNP_pos)),
                               columns =['pos', 'major', 'minor'])
```

```python
snp_pos_list=[]
major_list=[]
minor_list=[]

for snp in masked_major_minor_alleles.index:
    
    snp_info=masked_major_minor_alleles.loc[snp,:]
    
    for minor_allele in snp_info.minor:
        snp_pos_list.append(snp_info.pos)
        major_list.append(snp_info.major)
        minor_list.append(minor_allele)
```

```python
masked_minor_alleles_info=pd.DataFrame()
masked_minor_alleles_info['pos']=snp_pos_list
masked_minor_alleles_info['major']=major_list
masked_minor_alleles_info['minor']=minor_list
masked_minor_alleles_info.head()
```

```python
np.shape(masked_minor_alleles_info)
```

```python
pickle.dump( masked_minor_alleles_info, open( "../vars/masked_minor_alleles_info_treemmer_filtered", "wb" ) )
```

```python
# Count the number of events in the mutation_events table matching each position and minor allele
n_mutation_events_per_mutation = Counter([str(mut_i_pos) + '_' + gene + '_' + mut_i_der_call for mut_i_pos, gene, mut_i_der_call in zip(recombFree_mutation_events.Position , recombFree_mutation_events.Gene, recombFree_mutation_events.Derived_Call)])
```

```python
homoplasy_count = []
for SNP_i_pos , SNP_i_gene, SNP_i_minor_allele in zip(t.pos , t.Gene, t.minor):
        
    #count the number of mutational events that match this SNP position and minor allele
    SNP_i_minor_allele_count = n_mutation_events_per_mutation[str(SNP_i_pos) + '_' + SNP_i_gene + '_' + SNP_i_minor_allele]
    homoplasy_count.append(SNP_i_minor_allele_count)
```

### 3. Sort mutations by their enrichment in clusters

```python
homoplasy_info_annotated=pickle.load(open( "../vars/20210628_homoplasy_info_annotated", "rb" ) )
```

Take all SNPs where homoplasy count is at least 2:

```python
homoplasies=homoplasy_info_annotated[homoplasy_info_annotated.homoplasy_count>1]
np.shape(homoplasies)
```

Get a dictionary mapping the snp position to the flush-alignment position (full, flush-alignment position is indexed from 0:

```python
snp_pos=[line.rstrip('\n') for line in open('/n/data1/hms/dbmi/farhat/nikki/abscessus/0_NOTEBOOKS/010_homoplasy/vars/mab_masked_snp_positions.txt')]
snp_pos_series=pd.Series(snp_pos)
snp_pos_dict=dict(zip(snp_pos_series, snp_pos_series.index))
```

Load in the full alignment:

```python
masked_snpAln_unwrapped="/n/data1/hms/dbmi/farhat/nikki/abscessus/0_NOTEBOOKS/010_homoplasy/vars/mab_masked_snpAln_unwrapped.fasta"
aln_masked=AlignIO.read(masked_snpAln_unwrapped, "fasta")
```

For each SNP position, calculate its enrichment in the clusters:

```python
# isolates in each of the three major clusters:
clustA=[line.rstrip('\n') for line in open('/n/data1/hms/dbmi/farhat/nikki/abscessus/metadata/mab_clusterA.txt')]
clustB=[line.rstrip('\n') for line in open('/n/data1/hms/dbmi/farhat/nikki/abscessus/metadata/mab_clusterB.txt')]
clustC=[line.rstrip('\n') for line in open('/n/data1/hms/dbmi/farhat/nikki/abscessus/metadata/mab_clusterC.txt')]
```

### [3A]. enrichment in all 3 clusters:

```python
all_clust=clustA+clustB+clustC
```

```python
dominant_clust_enrichment={}
for pos, minor, major in zip(homoplasy_info_annotated.Position, homoplasy_info_annotated.minor, homoplasy_info_annotated.major):
    snp_position=snp_pos_dict[str(pos)]
    minor_allele_isolates=[i.id for i in aln_masked if i.seq[snp_position]==minor]
    major_allele_isolates=[i.id for i in aln_masked if i.seq[snp_position]==major]
    
    a=len(set(minor_allele_isolates).intersection(set(all_clust)))
    b=len(minor_allele_isolates)-a
    c=len(set(major_allele_isolates).intersection(set(all_clust)))
    d=len(major_allele_isolates)-c
    
    oddsr,p=fisher_exact([[a,b],[c,d]])
    
    dominant_clust_enrichment[pos+'_'+minor]=[oddsr,p]
```

```python
p_vals = []
for pos, minor, major in zip(homoplasy_info_annotated.Position, homoplasy_info_annotated.minor, homoplasy_info_annotated.major):
        
    #count the number of mutational events that match this SNP position and minor allele
    p_val_i = dominant_clust_enrichment[pos+'_'+minor][1]
    p_vals.append(p_val_i)
    
```

```python
odds_rats = []
for pos, minor, major in zip(homoplasy_info_annotated.Position, homoplasy_info_annotated.minor, homoplasy_info_annotated.major):
        
    #count the number of mutational events that match this SNP position and minor allele
    OR_i = dominant_clust_enrichment[pos+'_'+minor][0]
    odds_rats.append(OR_i)
```

```python
homoplasy_dominant_clust_enrich=homoplasy_info_annotated
homoplasy_dominant_clust_enrich['enrichment_p']=p_vals
homoplasy_dominant_clust_enrich['OR']=odds_rats
```
