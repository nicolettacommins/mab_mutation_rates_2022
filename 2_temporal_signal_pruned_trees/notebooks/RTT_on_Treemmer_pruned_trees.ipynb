```python
pg.corr(x=t95_data['date'], y=t95_data['distance'])
```

```python
X = t95_data['date'].values.reshape(-1,1)
y = t95_data['distance'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(11, 6), dpi=100)
plt.scatter(
    X,
    y,
    facecolors='none',
    edgecolors='black',
    s=250
)
plt.plot(
    X,
    predictions,
    c='lightseagreen', 
    linewidth=2,
)

new_rc_params = {'text.usetex': False,
"svg.fonttype": 'none'
}
matplotlib.rcParams.update(new_rc_params)

plt.xlabel("Date of Collection", fontsize=30, fontname="Arial", labelpad=15)
plt.ylabel("Root-to-tip distance", fontsize=30, fontname="Arial", labelpad=15)
plt.xticks(fontsize=22, fontname="Arial")
plt.yticks(fontsize=22, fontname="Arial")
sns.despine()
```

### Now let's randomly permute the dates and assess an empirical p-value:

```python
import random
X1=list(t95_data['date'].values)
Y=list(t95_data['distance'].values)
corr=[]
for i in range(0,1000):
    X2=random.sample(X1, len(X1))
    corr.append(pg.corr(X2, Y).r[0])
```

```python
plt.hist(corr, bins=30)
plt.ylabel('Frequency', fontsize=18)
plt.xlabel('Correlation Coefficient', fontsize=18)
plt.axvline(x=0.39, color='orange')

plt.savefig('RTT_permutation.svg', dpi=200, format='svg')
```

```python
# Because this is random, the output will be very slightly different every time.
sum(np.array(corr)>0.39)/1000
```
